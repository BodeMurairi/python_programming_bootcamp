### Unemployement Benefits Search vs. Actual Unemployment in the U.S.

**Challenge** Plot the search for "unemployment benefits" against the unemployment rate. 
1. Change the title to: Monthly Search of "Unemployment Benefits" in the U.S. vs the U/E Rate <br>
2. Change the y-axis label to: FRED U/E Rate <br>
3. Change the axis limits <br>
4. Add a grey [grid](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.grid.html) to the chart to better see the years and the U/E rate values. Use dashes for the line style<br> 
5. Can you discern any seasonality in the searches? Is there a pattern? 

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Assuming you have your data in df_unemployment

# Define locators and formatter for the x-axis
years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y')

# Create the figure with a larger size and higher resolution
plt.figure(figsize=(16, 10), dpi=150)

# Set an informative title with larger font size
plt.title('Monthly Search Trend of "Unemployment Benefits" in the U.S. Compared to U.E. Rate (2022-2023)',
          fontsize=20)

# Adjust x-axis tick labels for readability with rotation
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)

# Create twin axes and set labels with colors and font size
ax1 = plt.gca()
ax2 = ax1.twinx()
ax1.set_ylabel('FRED U/E Rate (%)', color='purple', fontsize=16)
ax2.set_ylabel('Search Trend (Normalized)', color='skyblue', fontsize=16)

# Set locators and formatter for the x-axis
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

# Set limits for the y-axes with some padding
ax1.set_ylim(bottom=2.5, top=11)  # Adjust limits as needed for your data
ax2.set_ylim(bottom=0, top=ax2.get_ylim()[1] * 1.1)  # Set upper limit based on existing data, add 10% padding

# Add grid lines with a subtle style
ax1.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.7)

# Plot the unemployment rate with dashed line style and thicker line width
ax1.plot(df_unemployment.MONTH, df_unemployment.UNRATE, color='purple', linewidth=2, linestyle='--')

# Plot the search trend with a different color and solid line style
ax2.plot(df_unemployment.MONTH, df_unemployment.UE_BENEFITS_WEB_SEARCH / df_unemployment.UE_BENEFITS_WEB_SEARCH.max(),
        color='skyblue', linewidth=2)  # Normalize search trend for better comparison

plt.show()
```

**Challenge**: Calculate the 3-month or 6-month rolling average for the web searches. Plot the 6-month rolling average search data against the actual unemployment. What do you see in the chart? Which line moves first?


```python
roll_df = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates  # Corrected import

# Assuming you have your data in df_unemployment

# Define locators and formatter for the x-axis (if not already defined)
years = mdates.YearLocator()
months = mdates.MonthLocator()
years_fmt = mdates.DateFormatter('%Y-%b')  # Format with year and abbreviated month

# Create the figure with a larger size and higher resolution
plt.figure(figsize=(16, 10), dpi=150)

# Set an informative title with larger font size
plt.title('Rolling 6-Month Average of U.S. "Unemployment Benefits" Web Searches vs. U.E. Rate (2022-2023)',
          fontsize=20)

# Adjust x-axis tick labels for readability with rotation
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)

# Create twin axes and set labels with colors and font size
ax1 = plt.gca()
ax2 = ax1.twinx()
ax1.set_ylabel('FRED U/E Rate (%)', color='purple', fontsize=16)
ax2.set_ylabel('Search Trend (Rolling Average)', color='skyblue', fontsize=16)

# Set locators and formatter for the x-axis (if not already set)
ax1.xaxis.set_major_locator(years)
ax1.xaxis.set_major_formatter(years_fmt)
ax1.xaxis.set_minor_locator(months)

# Set limits for the y-axes with some padding
ax1.set_ylim(bottom=2.5, top=11)  # Adjust limits as needed for your data
ax2.set_ylim(bottom=0, top=ax2.get_ylim()[1] * 1.1)  # Set upper limit based on existing data, add 10% padding

# Add grid lines with a subtle style
ax1.grid(color='grey', linestyle='--', linewidth=0.5, alpha=0.7)

# Calculate the rolling average over a 6-month window
roll_df = df_unemployment[['UE_BENEFITS_WEB_SEARCH', 'UNRATE']].rolling(window=6).mean()

# Plot the unemployment rate with a dashed line style and thicker line width
ax1.plot(df_unemployment.MONTH, roll_df.UNRATE, color='purple', linewidth=2, linestyle='--', label='U.E. Rate (Rolling Average)')

# Plot the search trend with a different color and solid line style
ax2.plot(df_unemployment.MONTH, roll_df.UE_BENEFITS_WEB_SEARCH, color='skyblue', linewidth=2, label='Search Trend (Rolling Average)')

# Add legends to distinguish between the lines
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize=14)

plt.show()
```
